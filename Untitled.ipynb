```python
# txt_path = "gpu_1_bs_32_20 (2).log"
txt_path = "nouseo.log"
lines = []
with open(txt_path, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        lines.append(line)
```

```python
import re
def get_metric(name, lines):
    p_list, r_list, f1_list = [], [], []
    for line in lines:
        if name in line:
            p = re.findall(r"Precision: (.*), Recall: ", line)
            if len(p) > 0:
                p_list.append(float(p[0]))
            r = re.findall(r"Recall: (.*), F1: ", line)
            if len(r) > 0:
                r_list.append(float(r[0]))
            f = re.findall(r"F1: (.*)", line)
            if len(f) > 0:
                f1_list.append(float(f[0]))
    return p_list, r_list, f1_list
            
bert1_result = get_metric("bert_crf_0", lines)
bert2_result = get_metric("bert_crf_1", lines)
bert_final_result = get_metric("final_bert_crf", lines)
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from matplotlib.pyplot import MultipleLocator


def draw_one(y1, y2, y3,name):
    y1 = y1[-len(y3):]
    y2 = y2[-len(y3):]
    plt.plot(y1[29:][::30],c='red')
    plt.plot(y2[29:][::30],c='green')
    plt.plot(y3[29:][::30],c='blue')
    
    # 
    ax = plt.gca()
    x_major_locator=MultipleLocator(1)
    ax.xaxis.set_major_locator(x_major_locator)
    
    y_major_locator=MultipleLocator(1)
    ax.yaxis.set_major_locator(y_major_locator)
    plt.xlabel(name ,fontsize=14)
    
    plt.grid()
    plt.show()

draw_one(bert1_result[0], bert2_result[0],bert_final_result[0], "precision")
draw_one(bert1_result[1], bert2_result[1],bert_final_result[1], "recall")
draw_one(bert1_result[2], bert2_result[2],bert_final_result[2], "f1")

draw_one(bert_final_result[0], bert_final_result[1],bert_final_result[2], "final")
# draw_one(bert2_result[0], "bert 2 precision")
# draw_one(bert_final_result[0],"bert final precision")

# draw_one(bert1_result[1], "bert 1 recall")
# draw_one(bert2_result[1], "bert 2 recall")
# draw_one(bert_final_result[1],"bert final recall")

# draw_one(bert1_result[2], "bert 1 f1")
# draw_one(bert2_result[2], "bert 2 f1")
# draw_one(bert_final_result[2],"bert final f1")
```
